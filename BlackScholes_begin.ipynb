```python
import numpy as np
from scipy.stats import norm
```

```python
#parameters

S = 30
K = 40
r = 0.01
T = 240/365
sigma = 0.30
```

```python
def blackscholes(r,S,K,T,type="C"):
  #calculate black scholes option price for call and put

  d1 = (np.log(S/K)+(r+sigma**2/2)*T)/(sigma*np.sqrt(T))
  d2 = d1 - sigma*np.sqrt(T)

  try:
    if type == "C":
      price = S*norm.cdf(d1, 0, 1) - K*np.exp(-r*T)*norm.cdf(d2, 0, 1)

    elif type == "P":
      price = K*np.exp(-r*T)*norm.cdf(-d2, 0, 1) - S*norm.cdf(-d1, 0, 1)
    return price
  except:
    print("Please confirm the option parameters input above.")

print("Option price is: ", round(blackscholes(r,S,K,T,type="C"), 2))
```
